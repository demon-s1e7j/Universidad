```python
#%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib as mpl
import matplotlib.animation as animation
from tqdm import tqdm
```

```python
Nt = 1000
Nx = 20
Ny = 20

x = np.linspace(0,2,Nx)
y = np.linspace(0,2,Nx)
t = np.linspace(0,100,Nt)

deltax = x[1] - x[0]
deltay = y[1] - y[0]
deltat = t[1] - t[0]

k = 0.0075

alpha = 1.
beta = 1.
omega = 0.1/4

lambda_ = (k*deltat)/deltax**2
mu_ = (k*deltat)/deltay**2

print(lambda_,mu_)
```

```python
def initT():
    T = np.zeros((len(t),len(x),len(y)))
    for l in range(len(t)):
        T[l,:,0] = np.e**(-2*t[l]*np.pi**2)*np.sin(np.pi*y)
        T[l,0,:] = np.e**(-2*t[l]*np.pi**2)*np.sin(np.pi*x)
        T[l,:,-1] = np.e**(-2*t[l]*np.pi**2)*np.sin(np.pi*(1+y))
        T[l,-1,:] = np.e**(-2*t[l]*np.pi**2)*np.sin(np.pi*(1+x))
    
    return T
```

```python
T = initT()
#print(T[0,:,:])
```

```python
def GetSolution():
    
    for l in tqdm(range(1,len(t))):
        
        #T[l,0,:] = np.sin(20*t[l])
        
        for i in range(1,len(x)-1):
            for j in range(1,len(y)-1):
                T[l,i,j] = (1-2*lambda_-2*mu_)*T[l-1,i,j] + \
                lambda_*(T[l-1,i+1,j] + T[l-1,i-1,j]) + \
                mu_*(T[l-1,i,j+1] + T[l-1,i,j-1])
```
