# Librerias a Utilizar

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Definici贸n de variables y soluci贸n de ecuaciones
## Variables
* t : Variable Independiente
* Y : Gamma (Desface)
* w : Omega (Velocidad Angular)
* A : Amplitud
* k : Constante del resorte
* m : Masa
* x : Posici贸n
* v : Velocidad
* E : Energia Total (constante)
* E_k : Energia Cinetica
* E_p : Energia Potencial

```python
t = np.linspace(0,10,1000)
Y = -np.pi/2
"""
Dadas las condiciones iniciales X0 = 0 y V0 = 0.1 entonces sabemos que  Y debe ser un numero tal que cumpla con
x = 0 y v = 0 cuando t es 0 y como X0 es descrito por x = Acos(wt+Y). Este solo puede ser pi/2 o -pi/2.
Sin embargo, como V = 1 y V = -AwSin(wt+Y) entonces Y = -pi/2
"""
A = 0.1/np.sqrt(5)
"""
Si tenemos las ecuaciones en t=0 podemos hallar w y A
"""
w = np.sqrt(5)
k = 0.5
m = 0.1
x = A*np.cos(w*t+Y)
v = -w*A*np.sin(w*t+Y)
E = (1/2)*(A**2)*k
E_k = (1/2)*(m*v**2)
E_p = E - E_k
```

# Graficas

## Posici贸n Vs Tiempo

```python
plt.plot(t,x)
```

## Velocidad Vs Tiempo

```python
plt.plot(t,v)
```

## Energia Cinetica Vs Tiempo

```python
plt.plot(t,E_k)
```

## Energia Potencial Vs Tiempo

```python
plt.plot(t,E_p)
```
