```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Datos_Hidrogeno = pd.read_csv("../Datos-Hidrogeno.csv")
X, Y = (Datos_Hidrogeno['Distance_(cm)'],Datos_Hidrogeno['Gray_Value'])
Datos_Hidrogeno
```

```python
np.polyfit(X,Y,1)
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y,color="#d65d0e")
plt.xlabel("Distancia (cm)")
plt.ylabel("Intensidad")
plt.savefig("Hidrogeno.png")
```

```python
Datos_Helio = pd.read_csv("../Datos-Helio.csv")
X, Y = (Datos_Helio['Distance_(cm)'],Datos_Helio['Gray_Value'])
Datos_Helio
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y,color="#98971a")
plt.xlabel("Distancia (cm)")
plt.ylabel("Intensidad")
plt.savefig("Helio.png")
```

```python
Y = np.array([434,486,656])
X = np.array([4.6,5.4,6.8])
poly = np.polyfit(X,Y,1)
adjust = np.poly1d(poly)
np.polyfit(X,Y,1,cov="unscaled")
```

```python
fig, ax = plt.subplots()
ax.plot(X,Y,"bo")
ax.plot(X,adjust(X),color="tomato")
plt.xlabel("Distancia (cm)")
plt.ylabel("Longitud de Onda (nm)")
plt.savefig("Flambda.png")
```

```python
#Y = np.array([adjust(4.5),adjust(5.0),adjust(6.0),adjust(6.8)])
Y = np.array([adjust(7.4),adjust(6.3),adjust(5.4),adjust(4.8)])
X = np.array([3,4,5,6])
poly2 = np.polyfit(1/X**2,1/Y,1)
adjust2 = np.poly1d(poly2)
np.polyfit(1/X**2,1/Y,1,cov=True)
```

```python
fig, ax = plt.subplots()
ax.plot(1/X**2,1/Y,"bo")
ax.plot(1/X**2,adjust2(1/X**2),color="tomato")
plt.savefig("grafica_mr.png")
```
