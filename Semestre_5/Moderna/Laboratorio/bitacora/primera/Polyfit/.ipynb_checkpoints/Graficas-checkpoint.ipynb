```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Datos = pd.read_csv('Datos_Exp1.csv')
Azul = Datos[Datos['Color'] == 'Azul']
Amarillo = Datos[Datos['Color'] == 'Amarillo']
Verde = Datos[Datos['Color'] == 'Verde']
Rojo = Datos[Datos['Color'] == 'Rojo']
nums = np.linspace(1,5,1000)

plt.plot(Amarillo['Corriente'],Amarillo['Voltaje'],'yo')
plt.plot(Azul['Corriente'],Azul['Voltaje'],'bo')
plt.plot(Verde['Corriente'],Verde['Voltaje'],'go')
plt.plot(Rojo['Corriente'],Rojo['Voltaje'],'ro')
plt.legend(('Amarillo','Azul','Verde','Rojo'))
plt.title('Grafia de Corriente vs Voltaje')
plt.xlabel('Corriente')
plt.ylabel('Voltaje')
plt.show()
```

```python
Amarillo
```

```python
XA1 = Amarillo['Corriente'].to_numpy()
YA1 = Amarillo['Voltaje'].to_numpy()
YA1 = YA1.astype(np.float64)
adjustA1 = np.polyfit(XA1,YA1,1)
polyA1=np.poly1d(adjustA1)
np.polyfit(XA1,YA1,1,cov=True)
```

```python
fig, ax = plt.subplots()
ax.errorbar(XA1,YA1,yerr=0.1,xerr=0.05,fmt="yo")
ax.plot(XA1,polyA1(XA1),color="Tomato")
plt.title("Grafica Amarillo")
plt.xlabel("Corriente")
plt.ylabel("Voltaje")
plt.show
```

```python
X = Azul['Corriente'].to_numpy()
Y = Azul['Voltaje'].to_numpy()
Y = Y.astype(np.float64)
adjust = np.polyfit(X,Y,1)
poly=np.poly1d(adjust)
np.polyfit(X,Y,1,cov=True)
```

```python
fig, ax = plt.subplots()
ax.errorbar(X,Y,yerr=0.005,fmt="bo")
ax.plot(X,poly(X),color="Tomato")
plt.title("Grafica Azul")
plt.xlabel("Corriente")
plt.ylabel("Voltaje")
plt.show
```

```python
X = np.array([4.55,5.29,6.40])
Y = np.array([0.315,0.322,0.718])
X = X*10**(14)
Y2 = Y*(1.6)/10**19
adjust = np.polyfit(X,Y2,1)
poly=np.poly1d(adjust)
np.polyfit(X,Y2,1,cov=True)
```

```python
fig, ax = plt.subplots()
ax.errorbar(X,Y2,fmt="go")
ax.plot(X,poly(X),color="Tomato")
plt.title("Grafica Frecuencia vs Kmax")
plt.xlabel("Frecuencia")
plt.ylabel("Kmax")
plt.show
```
