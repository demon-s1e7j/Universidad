# Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Datos

```python
H_Ne = pd.read_csv("H_Ne.csv")
H = pd.read_csv("H.csv")
D = pd.read_csv("D.csv")
```

# Graficas Iniciales

## Grafica Hidrogeno Neon

```python
X = H_Ne["Pixel"].to_numpy()
Y = H_Ne["Brillo"].to_numpy()
plt.plot(X,Y,color="tomato")
plt.xlabel("Pixel")
plt.ylabel("Brillo")
plt.title("Distribuci√≥n del Brillo entre Hidrogeno y Neon")
plt.savefig("H_Ne.png")
```

```python
X = H["Pixel"].to_numpy()
Y = H["Brillo"].to_numpy()
plt.plot(X,Y,color="tomato")
plt.xlabel("Pixel")
plt.ylabel("Brillo")
plt.title("Distribuci√≥n del Brillo de Hidrogeno")
plt.savefig("H.png")
```

```python
X = D["Pixel"].to_numpy()
Y = D["Brillo"].to_numpy()
plt.plot(X,Y,color="tomato")
plt.xlabel("Pixel")
plt.ylabel("Brillo")
plt.title("Distribuci√≥n del Brillo de Deuterio")
plt.savefig("D.png")
```

# Poligono

```python
X = [42,1068]
Y = [653.288,656.27097]
poly= np.poly1d(np.polyfit(X,Y,1))
fig, ax = plt.subplots()
ax.plot(X,Y,"ro")
ax.plot(X,poly(X))
plt.xlabel("Pixel")
plt.ylabel("Longitud de Onda")
plt.title("Relaci√≥n entre Pixeles y Longitud de Onda para Hidrogeno/Neon")
plt.savefig("Polygono.png")
```

# Estructura Fina Hidrogeno

```python
# Los puntos maximos del hidrogeno estan en
X = np.array([23,26])
# Por lo tanto, la "Constante de Estructura Fina es"
ConstanteEstructuraFinaHidrogeno = poly(X[1])-poly(X[0])
ConstanteEstructuraFinaHidrogeno
```
