# Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Datos

```python
Hg = pd.read_csv("Hg.csv")
```

# Grafica Inicial

```python
X = Hg["Pixel"].to_numpy()
Y = Hg["Brillo"].to_numpy()
plt.plot(X,Y,color="tomato")
plt.xlabel("Pixel")
plt.ylabel("Brillo")
plt.title("Distribuci√≥n del Brillo entre Hidrogeno y Neon")
plt.savefig("Hg.png")
```

# Poligono

```python
X = [54,696]
Y = [576.9610,579.0670]
poly= np.poly1d(np.polyfit(X,Y,1))
fig, ax = plt.subplots()
ax.plot(X,Y,"ro")
ax.plot(X,poly(X))
plt.xlabel("Pixel")
plt.ylabel("Longitud de Onda")
plt.title("Relaci√≥n entre Pixeles y Longitud de Onda para el Mercurio")
plt.savefig("PolygonoHg.png")
```
