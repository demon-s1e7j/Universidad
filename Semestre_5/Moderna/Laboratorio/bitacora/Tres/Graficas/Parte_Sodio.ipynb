# Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Datos

```python
Na = pd.read_csv("Na.csv")
```

# Grafica Inicial

```python
X = Na["Pixel"].to_numpy()
Y = Na["Brillo"].to_numpy()
plt.plot(X,Y,color="tomato")
plt.xlabel("Pixel")
plt.ylabel("Brillo")
plt.title("Distribución del Brillo para el Sodio")
plt.savefig("Na.png")
```

# Poligono

```python
X = [66,250]
Y = [285.3013,289.3618]
poly= np.poly1d(np.polyfit(X,Y,1))
fig, ax = plt.subplots()
ax.plot(X,Y,"ro")
ax.plot(X,poly(X))
plt.xlabel("Pixel")
plt.ylabel("Longitud de Onda")
plt.title("Relación entre Pixeles y Longitud de Onda para el Sodio")
plt.savefig("PolygonoNa.png")
```

# Estructura Fina Sodio 1

```python
# Los puntos maximos del hidrogeno estan en
X = np.array([64,67])
# Por lo tanto, la "Constante de Estructura Fina es"
ConstanteEstructuraFinaSodio1 = poly(X[1])-poly(X[0])
ConstanteEstructuraFinaSodio1
```
