# Librerias

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Actividad 1

```python
datosActividad1 = pd.read_csv("Datos_Campo.csv")
X = datosActividad1["Corriente (A)"]
Y = datosActividad1["Campo Magnetico (mT)"]
polyActividad1 = np.polyfit(X,Y,1)
adjustActividad1 = np.poly1d(polyActividad1)
fig, ax = plt.subplots()
ax.plot(X,Y,"go")
ax.plot(X,adjustActividad1(X))
plt.title("Grafica de Campo Magnetico en Funci√≥n de la Corriente")
plt.xlabel("Corriente (A)")
plt.ylabel("Campo Magnetico (mT)")
plt.savefig("Actividad1.png")
print(np.polyfit(X,Y,1,cov="unscaled"))
```

# Actividad 2

```python
datosActividad2 = pd.read_csv("Datos.csv")
voltajes = np.array([280,261,250,236,224,205,190,181,170,160,150,140])
fig, ax = plt.subplots()
for voltaje in voltajes:
    datos = datosActividad2[datosActividad2["Voltaje (V)"]==voltaje]
    X = datos["Distancia"]
    Y = datos["Corriente (A)"]
    ax.plot(X/2,Y,label=str(voltaje)+" V")
plt.legend()
plt.title("Grafica de Corriente en Funci√≥n de la Distancia")
plt.ylabel("Corriente (A)")
plt.xlabel("Distancia (cm)")
plt.savefig("Actividad21.png")
```
